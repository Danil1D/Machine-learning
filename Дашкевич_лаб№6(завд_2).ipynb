```python
# Завантаження необхідних бібліотек для кластеризації
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
# Створення штучних даних для кластеризації
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Використаємо метод ліктя для визначення оптимальної кількості кластерів
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Побудова графіку ліктя
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.title('Метод ліктя')
plt.show()

# Використаємо метод силуетів для вибору кількості кластерів
silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# Побудова графіку для методу силуетів
plt.figure(figsize=(8, 5))
plt.plot(range(2, 10), silhouette_scores, 'bo-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Silhouette Score')
plt.title('Метод силуетів')
plt.show()

# Кластеризація з оптимальною кількістю кластерів (наприклад, 4)
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Візуалізація кластерів
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Кластери та їх центри')
plt.show()
```

Висновок:

Базова лінійна модель показала ознаки перенавчання, а застосування регуляризації (Ridge і Lasso) значно знизило це перенавчання, покращивши загальну продуктивність моделей. Ridge забезпечив краще узагальнення, а Lasso допоміг скоротити незначущі ознаки.
